```python
import numpy as np
# Sending binary data via sinusoidal signal maniputation
```

```python
# Parameters
sig_len = 1000 # Signal length(in samples)
sampl_per_bin = 100 # Samples per binarty representation
bin_data_len = int(sig_len/sampl_per_bin)
bin_data = np.random.randint(low=0,high=2,size=bin_data_len)
```

```python
# Create sinusoidal carriers
sig_carrier_base = np.sin(2*np.pi*np.arange(0,1,(1/sampl_per_bin))) # Baseline carrier
sig_carrier_freq = np.sin(2*2*np.pi*np.arange(0,1,(1/sampl_per_bin))) # Double frequency
sig_carrier_phase = np.sin(2*np.pi*np.arange(0,1,(1/sampl_per_bin))+(np.pi/4)) # Phase shifted by 45 degreeands
```

```python
# Manipulations
sig_bin = np.array(()) # Binary waveform
sig_ask = np.array(()) # Amplitude modulated
sig_psk = np.array(()) # Phase modulated
sig_fsk = np.array(()) # Frequency modulated

for i in bin_data:
    if i == 1:
        sig_bin = np.append(sig_bin, np.ones((1,sampl_per_bin)))
        sig_ask = np.append(sig_ask, sig_carrier_base)
        sig_psk = np.append(sig_psk, sig_carrier_base)
        sig_fsk = np.append(sig_fsk, sig_carrier_base)
    else:
        sig_bin = np.append(sig_bin, np.zeros((1,sampl_per_bin)))
        sig_ask = np.append(sig_ask, 0.5*sig_carrier_base)
        sig_psk = np.append(sig_psk, sig_carrier_phase)
        sig_fsk = np.append(sig_fsk, sig_carrier_freq)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.subplot(411)
plt.plot(sig_bin)
plt.subplot(412)
plt.plot(sig_ask)
plt.subplot(413)
plt.plot(sig_fsk)
plt.subplot(414)
plt.plot(sig_psk)
plt.show()
```
