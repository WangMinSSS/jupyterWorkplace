# 数字射频存储（Digital Radio Frequency Memory）
### wikipedia
* 是一种在一定频率和带宽下对射频信号进行数字化捕获和重发的技术
* 要完整恢复捕获的信号
* 重发的信号要和原始信号保持相干
* 可以重塑信号然后重发产生假目标
* 最早的文献是1975年由Sheldon C.Spector写的《A Coherent Microwave Using Digital Storage:The Loopless Memory Loop》
* DRFM技术通过接收到对方的射频信号后保存一个相干的副本，在需要进行干扰的时候，重新发射出来。因为是雷达的原始信号的副本，而且相干，所以雷达无法区分重发的信号和真实的目标。在距离可以在我方目标前面或者后面产生假目标作为掩护，在速度上轻微的频率变化就可以让敌方雷达在多普勒检测上出错，在相位上的修改可以对敌方的单脉冲测角技术产生很大的影响。
* DRFM技术还可以用来做雷达系统的环路仿真，相比用实际的飞行检测，可以极大减少测试成本。

### 数字射频存储发展综述 —黄思源 《飞航导弹》2017
##### 摘要
结合数字射频存储技术在⼲扰雷达、导弹等⽅⾯发挥作⽤⽇益突出的现状,从DRFM的起源介绍了其发展历程,分析了⽬前DRFM的关键技术及DRFM在各领域尤其是军事领域的应⽤情况。最后研究DRFM的发展趋势,为进⼀步研究DRFM提供理论基础。
* DRFM技术以高速采样和数字存储技术为基础，通过存储、分析、复制一定频率和带宽的数字化射频信号，达到对雷达和导弹引信等设备进行干扰的目的。优点是：快速存储，保留特征信息，可实现多种干扰控制
* 存储之后，根据需要重构并调制假目标的速度、角度、距离、数量和功率等信息
* 可以制造大量虚假目标，或者用假目标诱使敌方隐藏雷达开机
* 国际上将噪声干扰称为第一代干扰技术，基于射频存储转发的干扰为第二代干扰
* 早期模拟射频存储（ARFM），主要有放大器和延迟线组成闭合回路实现。相干性差、设备体积质量大、灵活性差
* DRFM基本原理由英国EMI公司的Chris Haynes在1974年提出
* 衡量DRFM性能的主要技术参数包括工作频率范围、瞬时带宽、存储脉宽、储频精度、最小输入信号功率、杂散抑制、最小转发延迟、最大保存时间等
* 第三代DRFM（欧洲EADS公司，2001），覆盖瞬时带宽扩展到了1.8Ghz，频率范围扩展到了4Ghz，频率精度提升至1khz，功耗60w，同时处理的威胁数目达到四个
* 国内相关机构有：航空工业总公司601所、北京理工大学、西安电子科技大学、电子科技大学、中国科学技术大学、国防科技大学，重点研究基于DRFM的固定杂波、运动杂波、目标信号或噪声的控制。企业有华清瑞达、北京雷久科技、北京经纬润恒科技。最新产品技术参数：量化位数8—14位，瞬时带宽50Mhz到1.8Ghz，延迟精度<500ps，距离精度<0.075m，存储时间>4ms，存储带宽>4GB/s
* 在硬件技术、信号类型、批次数量、杂波模拟方面与国外还存在明显差距
* 结构类型分为：单通道、正交双通道和信道化DRFM
    * 正交双通道：通过改变两路信号的频率来对干扰信号附加不同的多普勒频移
    * 信道化：将DRFM的瞬时带宽分解为n个通道，每个通道的带宽为总带宽的1/n，使系统对A/D转换器件以及信号处理器件的要求大幅降低
* 对于DRFM而言，单独采用距离波门拖引或者速度拖引技术只能对只有距离或速度检测、跟踪能力的雷达奏效，无法欺骗具有距离－速度二维信息同时检测、跟踪能力的雷达或导弹。
* 现有DRFM干扰机产生的假目标大多是无序分布的，对雷达网难以达到理想的欺骗效果。
* DRFM可以用来进行电磁设备的硬件的在环仿真。

### 雷达欺骗干扰特征提取与综合态势感知方法综述—李艳莉 田晓 韦顺军
##### 摘要
雷达欺骗⼲扰的特征提取和综合感知是雷达抗⼲扰的前提条件,也是雷达抗欺骗⼲扰⼯作流程中的难点问题。从⼲扰机⽆意调制的细微特征、信号起伏特性、极化特征、动⼒学特性、微多普勒特征、组⽹雷达特性、多尺度分解特征、多域变换特性等⽅⾯综述了雷达欺骗⼲扰特征提取和综合感知⽅法的研究进展,总结了各种欺骗⼲扰感 知⽅法的约束条件、运算量和识别性能,最后,指出了⽬前雷达欺骗⼲扰的特征提取和综合感知技术存在的问题以及未来的发展⽅向。
##### 无意特征提取
* 随着DRFM的欺骗式干扰机在实际作战应用中大规模使用，雷达有源干扰主要使用样式为相参欺骗干扰
* DRFM从接收采样、上下变频到放大发射的各个环节均经过了信号的二次调制，在这些环节中均存在不同程度的谐波、杂散等非线性失真，可以针对这些细微的无意调制特征进行欺骗干扰的分析和识别
* DRFM**幅度和相位量化效应**，在进行数模转换的过程中存在量化误差，随着量化位数的增加而减小，目前该特征方法已不常用
* DRFM生成的距离拖引干扰（Range Gate Pull Off,RGPO）具有**时延离散过程**，会在雷达信号上叠加谐波，导致与真实回波产生均匀间隔的谐波频谱失真，从而使得中心频率产生偏移
* **利用欺骗干扰和雷达信号在本振相位噪声存在的指纹特征进行欺骗干扰的识别，**在DRFM干扰机中，本振与雷达发射机存在差异，欺骗干扰在经过上下变频的二次调制之后，信号在相位噪声上存在细微差异。**但这种方法对信噪比的要求较高**
* 欺骗式干扰机在功放放大过程中存在很大的非线性失真，可以通过**功放二次调制过程中附加的杂散特征进行假目标的识别，**是目前比较有效的技术途径
##### 综合感知方法
* DRFM干扰机转发的信号幅度和目标回波的幅度分布特性存在差异，此时可以将欺骗干扰的识别问题转换为存在欺骗干扰条件下的检测问题。目前**通过幅度起伏特性检测进行欺骗干扰识别**是一个研究的热点，但还未得到实测数据的验证
* **基于极化特性的感知方法，**目前DRFM干扰机一般采用固定极化方式，而雷达可以采用极化捷变、随机极化等，欺骗干扰和雷达目标信号在极化域存在较大差异
* **基于微多普勒特性的感知方法，**针对一些具有微动特征的目标，可以建立目标的微动多普勒模型进行特征提取和感知。微多普勒特征是弹道导弹的固有雷达特征。
* **基于组网雷达特性的感知方法，**在不同的散射方向，真实目标和有源干扰的雷达横截面（Radar Cross Section,RCS）的独立性存在差异，实际目标RCS将彼此独立，假目标的RCS在组网雷达中是相关的
* **基于动力学特性的感知方法，**对于弹道导弹防御系统的预警雷达，在数据处理层面，假目标很多情况下无法模拟真实目标的动力学特性
* **基于多尺度分解的感知方法，**欺骗干扰的实施过程就是在雷达回波信号上进行多维度调制，通过多尺度分解提取这些细微的调制特征可用于欺骗干扰的识别和调制参数的估计。该方法运算量较大
* **基于多域联合处理的感知方法，**在时域、频域、多普勒域等多个调制域提取联合特征参数进行欺骗干扰综合感知。该方法运算量较大

##### 存在的问题
* 欺骗⼲扰的准确建模问题。如何将各种影响因素综合在⼀起,构建和实际环境⾼度逼真的⼲扰模型,并通过实测试验数据对欺骗⼲扰的模型进⾏修正和完善,是后续⼲扰特征提取和综合感知的研究基础。
* ⼲扰感知算法在低信噪⽐条件下的适应性问题。要达到90%的识别概率,⽬前的算法对信噪比/干扰⽐的要求⼤多在4dB以上,特别是基于细微特征识别的信噪⽐/⼲扰⽐要求更是⼤于8dB,如何提⾼实际环境中的较低信噪⽐/干扰⽐的识别概率也是⼀个实际应⽤有较⼤挑战的研究⽅向。
* 新样式欺骗⼲扰的特征提取和综合感知⽅法算法更新问题。随着雷达技术体制的不断创新,欺骗⼲扰的样式也随之增加和完善,如何实现对新的欺骗⼲扰样式进⾏特征提取和综合识别将是该领域不断进⾏技术创新的持续研究⽅向。
