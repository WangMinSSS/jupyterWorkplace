```python
import numpy as np
import scipy as sci
import matplotlib.pyplot as plt
```

```python
n = [np.random.random() for i in range(10000)]
p = [] # 概率密度函数
c = [0] # 概率分布函数
for i in range(100):
    i = i/100
    m = len([x for x in n if x>=i and x<i+0.01])/10000
    p.append(m)
    c.append(m+c[-1])
plt.subplot('211')
plt.plot(p)
plt.subplot('212')
plt.plot(c)
plt.show()
```
