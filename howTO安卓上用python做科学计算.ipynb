## 安装python组件
经过多次失败的卸载重装，摸索发现scipy这个模块安装起来很费劲，我自己按一下步骤成功安装：

* 不要先安装Python，采用以下方式安装scipy库时它会安装依赖的python和numpy

   * 安装wget

pkg install wget

   * 下载文件
   
wget https://its-pointless.github.io/setup-pointless-repo.sh

   * 执行文件

bash setup-pointless-repo.sh

安装完成，自行测试

* 安装matplotlib画图库，'LDFLAGS="-lm -lcompiler_rt"'据说是在本地对库进行编译的意思

apt install libpng freetype pkg-config
LDFLAGS="-lm -lcompiler_rt" pip install matplotlib

* 安装jupyter

apt install libzmq
LDFLAGS="-lm -lcompiler_rt" pip install jupyter

* 安装jupyter主题

pip install  jupyterthemes

* 安装jupyter Nbextensions插件

pkg install libxml2 libxslt
pip install jupyter_contrib_nbextensions
jupyter contrib nbextension install --user # 激活插件

至此，所需组件安装完成

## 配置环境
主要是对jupyter notebook的主题以及插件进行配置

* jupyter环境配置
   * 终端输入，生成配置文件
   
jupyter notebook --generate-config

   * 进入目录打开配置文件
   
vim ~/.jupyter/jupyter_notebook_config.py

   * 在最下行添加如下代码

c.NotebookApp.ip = '*' # 开启所有的IP访问，即可使用远程访问
c.NotebookApp.open_browser = False # 关闭启动后的自动开启浏览器
c.NotebookApp.port = 8888  # 设置端口8888，也可用其他的，比如1080，8080等等
c,NotebookApp.notebook_dir = './filename' # 更改jupyter的工作目录

   * 设置密码，在终端输入
   
jupyter notebook password
   
* 主题选择

jt -l # 查看可选主题列表
jt -t grovboxl -T #这是我选择的主题，还可以对字体进行更改，自行查看jt命令帮助
